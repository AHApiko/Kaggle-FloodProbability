```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
plt.hist(train['FloodProbability'], bins=150)
plt.xlabel("FloodProbability")
plt.ylabel("Frequency")
plt.title('Distribution of Flood Probability')
```

```python
correlation_matrix = train.corr()
plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('correlation matrix')
plt.show
```

```python
from scipy.stats import spearmanr, kendalltau
for col in train.columns:
  if col != 'FloodProbability':
    corr, pval = spearmanr(train[col], train['FloodProbability'])
    print(f'Spearmans rank correlation between {col} and FloodPrability: {corr:.2f}, pvalue: {pval:.4f}')
```

```python
for col in train.columns:
  if col != 'FloodProbability':
    corr, pval = kendalltau(train[col], train['FloodProbability'])
    print(f'Spearmans rank correlation between {col} and FloodPrability: {corr:.2f}, pvalue: {pval:.4f}')
```
